```python
import numpy as np
import json
from collections import defaultdict
from pprint import pprint
```

```python
with open('assists_barcelona_raw.json') as f:
  assist_data = json.load(f)
```

```python
assist_data
```

```python
x_boundries = np.array([0, 20, 40, 60, 80, 100, 120])
y_boundries = np.array([0, 30, 50, 80])

zones = np.array([[ 1, 4, 7, 10, 13, 16],
                  [ 2, 5, 8, 11, 14, 17],
                  [ 3, 6, 9, 12, 15, 18]])

zones_x = np.array([0, 0, 0, 20, 20, 20, 40, 40, 40, 60, 60, 60, 80, 80, 80, 100, 100, 100])

zones_y = np.array([0, 30, 50, 0, 30, 50, 0, 30, 50, 0, 30, 50, 0, 30, 50, 0, 30, 50])

zones_w = np.array([20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20, 20])

zones_h = np.array([30, 20, 30, 30, 20, 30, 30, 20, 30, 30, 20, 30, 30, 20, 30, 30, 20, 30])

print(len(zones_x))
print(len(zones_y))
print(len(zones_w))
print(len(zones_h))
```

```python
assists_dict = defaultdict(list)
zones_dict = {}

for item in assist_data:
    season = item['season']
    season_assists_list = []
    season_zone_list = np.zeros(18)
    
    assists = item['assists']
    for assist in assists:
        if 'goal_assist' in assist:
            x, y = assist['start_location']
            
            if x > 120:
                x = 120
            
            if y > 80:
                y = 80
            
            row = np.digitize(y, y_boundries, right=True)-1
            column = np.digitize(x, x_boundries, right=True)-1
            zone = float(zones[row, column])
            
            cleaned_assist = {
                'start_x': x,
                'start_y': y,
                'zone': zone
            }
            
            season_zone_list[(int(zone-1))] += 1
            season_assists_list.append(cleaned_assist)
            
    assists_dict[season] = season_assists_list
    zones_dict[season] = season_zone_list
```

```python
assists_dict
```

```python
assists_json = json.dumps(assists_dict, indent=4, separators=(',', ':'))
```

```python
with open('assists_barcelona_clean.json', 'w') as json_file:
    json_file.write(assists_json),
json_file.close()
```

```python
zones_dict
```

```python
for key in zones_dict.keys():
    zones_list = []
    
    counts = zones_dict[key]
    for zone_idx in range(len(counts)):
        zone_data = {
            'zone': int(zone_idx+1),
            'count': int(counts[zone_idx]),
            'zone_x': int(zones_x[zone_idx]),
            'zone_y': int(zones_y[zone_idx]),
            'zone_w': int(zones_w[zone_idx]),
            'zone_h': int(zones_h[zone_idx])
        }
        zones_list.append(zone_data)
    
    zones_dict[key] = zones_list
```

```python
zones_dict
```

```python
zones_json = json.dumps(zones_dict, indent=4, separators=(',', ':'))

with open('assist_zones_barcelona_clean.json', 'w') as json_file:
    json_file.write(zones_json),
json_file.close()
```
